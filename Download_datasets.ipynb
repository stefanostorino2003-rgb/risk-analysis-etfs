```python
import yfinance as yf
import pandas as pd

# Lista ETF
tickers = ["SPY","QQQ","IWM","EFA","EEM","XLK","XLF","XLV","TLT","GLD"]

# Download dati (Adjusted Close = include dividendi)
data = yf.download(tickers, start="2005-01-01")["Close"]
```

```python
print(data.head())
```

```python
returns = data.pct_change().dropna()
print(returns.head())
```

```python
for ticker in returns.columns:
    returns[[ticker]].to_csv(f"{ticker}_returns.csv")
```

```python
tickers = ["SPY","QQQ","IWM","EFA","EEM","XLK","XLF","XLV","TLT","GLD"]

data = {}

for ticker in tickers:
    df = pd.read_csv(f"{ticker}_returns.csv", parse_dates=["Date"], index_col="Date")
    df.columns = ["Return"]
    data[ticker] = df
```
